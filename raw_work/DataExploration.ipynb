```python
con = sqlite3.connect("ratings.db")
cur = con.cursor()
```

```python
# Making raw columns match what I made in the db
data = data.rename(columns={
    "date": "published_at",
    "stock": "ticker",
    "Headline Class": "Headline_Class",
    "Unnamed: 0": "article_id"
})
data.head()
```

```python
# Remove the second article id 
data = data.iloc[:, 1:]
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.to_csv("cleaned_data.csv", index = False)
```

```python
# Actually put my data in the table from last week
data.to_sql("articles", con, if_exists="append", index=False)
con.commit()
```

```python
pd.read_sql("SELECT COUNT(*) AS n_rows FROM articles;", con)
```

```python
pd.read_sql("SELECT * FROM articles LIMIT 5;", con)
```

```python
cur.execute("SELECT COUNT(*) FROM articles;")
print(cur.fetchone())
```

```python
# Are the keys unique
cur.execute("""
            SELECT COUNT(*) AS n_rows,
            COUNT(DISTINCT article_id) AS n_distinct_ids
            FROM articles;
            """)
print(cur.fetchone())
```
